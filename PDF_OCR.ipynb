# Text Filter

```python
def search_kupfer(filepath):
  with open(filepath,'r') as file:
    #Read file in memory
    #if too slow, try enumerate (see: https://pynative.com/python-search-for-a-string-in-text-files/)
    content = file.read()

    #Specify what you want entities you want to look for (Kupfer)
    word = 'Kupfer Kabel'
    name_end = 'Cu'
    charac_after_word = 30
    #Specify where Einheitspreis will start
    character_after_word = 40
    #Look for TO, 2 characters after the Einheitspreis will start
    tonnage = 'TO'

    #initiate list for entity name (Kupfergehalt)
    kupfer_list = []
    #initiate list for Einheitspreis
    kupfer_EH = []
    #initiate list for date
    kupfer_date = []

    

    for m in re.finditer(word,content):
      #print("Kupfer Kabel found at: ", m.end())
      for i in re.finditer(name_end,content[m.end():m.end()+charac_after_word]):
        kupfer_name = content[m.start():m.end()+i.end()]
        #print("Whole name: ",kupfer_name)
        kupfer_list.append(kupfer_name)
      for n in re.finditer(tonnage,content[m.end():m.end()+character_after_word]):
        #print("TO found at: ", n.end())

        #Conversion from string to float
        kupfer_einhp = content[m.end()+n.end()+1:m.end()+n.end()+9]
        kupfer_einhp = kupfer_einhp.replace(".","")
        kupfer_einhp = kupfer_einhp.replace(",",".")
        kupfer_einhp = float(kupfer_einhp)
        #print('Einheitspreis: ', kupfer_einhp)

        kupfer_EH.append(kupfer_einhp)

  return kupfer_list, kupfer_EH, kupfer_date
```

```python
for filename in file_list:
  #inititate empty lists
  kupfer_list = []
  kupfer_EH = []
  kupfer_date = []

  #search for data in file
  output_txt_filename = output_txt_folder_path+'/'+filename.replace("pdf","txt")
  kupfer_list_new, kupfer_EH_new, kupfer_date_new = search_kupfer(output_txt_filename)

  #write it to list
  kupfer_list.append(kupfer_list_new)
  kupfer_EH.append(kupfer_EH_new)
  kupfer_date.append(kupfer_date_new)

print(kupfer_list,kupfer_EH, kupfer_date)
```
